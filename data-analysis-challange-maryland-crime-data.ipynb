```python
import pandas as pd

# Path to the CSV file
file_path = '/kaggle/input/md-crime/MD_Crime_Data.csv'

# Read the CSV file
crime_data = pd.read_csv(file_path)


# Separate the 'JURISDICTION' column
jurisdiction_column = crime_data['JURISDICTION']

# Drop the 'JURISDICTION' column from the dataframe
crime_data = crime_data.drop(columns=['JURISDICTION'])

# Replace missing values with the mean of their respective columns
crime_data.fillna(0, inplace=True)

# Add the 'JURISDICTION' column back to the dataframe
crime_data['JURISDICTION'] = jurisdiction_column


# Check for NaN values to confirm they have been replaced
print("\nCheck for NaN values in the dataset after replacement:")
print(crime_data.isna().sum())
```

# 1.Trend Analysis: How have the overall crime rates in Maryland changed from 1975 to 2020? Identify any significant upward or downward trends.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Separate the 'JURISDICTION' column
jurisdiction_column = crime_data['JURISDICTION']

# Drop the 'JURISDICTION' column from the dataframe
crime_data = crime_data.drop(columns=['JURISDICTION'])

# Replace missing values with the mean of their respective columns
crime_data.fillna(crime_data.mean(), inplace=True)

# Add the 'JURISDICTION' column back to the dataframe
crime_data['JURISDICTION'] = jurisdiction_column

# Group by year and calculate the sum of 'GRAND TOTAL' for each year
crime_data_by_year = crime_data.groupby('YEAR').sum()

# Plot the trend of overall crime rates over the years with all years on the x-axis
plt.figure(figsize=(14, 8))
plt.plot(crime_data_by_year.index, crime_data_by_year['GRAND TOTAL'], marker='o')
plt.title('Overall Crime Rates in Maryland (1975-2020)')
plt.xlabel('Year')
plt.ylabel('Total Crime Incidents')
plt.xticks(crime_data_by_year.index, rotation=90)  # Rotate x-axis labels for better readability
plt.grid(True)
plt.tight_layout()  # Adjust layout to ensure everything fits without overlap
plt.show()
```

Significant Trends:

Upward Trend (1975-1994): This period saw a significant increase in crime rates, particularly noticeable in the late 1980s.

Downward Trend (1995-2000): After peaking in 1995, there was a consistent decrease in crime rates through the 1990s.

Stable Period (2001-2020): Crime rates generally stabilized or continued to decrease slightly, reflecting broader societal changes and possibly improved crime prevention measures.

# 2. Crime Distribution: What are the most common types of crimes committed in Maryland over the years? How has the distribution of different crime types changed over time?

```python
import pandas as pd
import matplotlib.pyplot as plt

# You can adjust these columns based on your specific dataset
crime_categories = ['MURDER', 'RAPE', 'ROBBERY', 'AGG. ASSAULT', 'B & E', 'LARCENY THEFT', 'M/V THEFT']
crime_data_by_year = crime_data.groupby('YEAR')[crime_categories].sum()

# Plotting the trends for each crime category
plt.figure(figsize=(14, 10))

for crime_type in crime_categories:
    plt.plot(crime_data_by_year.index, crime_data_by_year[crime_type], marker='*', label=crime_type)

plt.title('Maryland Crime Trends by Type (1975-2020)')
plt.xlabel('Year')
plt.ylabel('Number of Incidents')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.xticks(crime_data_by_year.index, rotation=90)  # Rotate x-axis labels for better readability
plt.show()
```

From the provided data on crime types in Maryland from 1975 to 2020, we can infer the following regarding the most common types of crimes and how their distribution has changed over time:

### Most Common Types of Crimes:
1. **Larceny Theft**: Throughout the entire period, Larceny Theft consistently appears as one of the most common types of crimes, with high numbers reported each year.
   
2. **Burglary (B & E)**: Similar to Larceny Theft, Burglary (B & E) also shows relatively high numbers consistently across the years, although there are fluctuations.

3. **Aggravated Assault**: This crime type also shows significant numbers consistently over the years, indicating its prevalence in Maryland.

### Changes in Crime Distribution Over Time:
- **Murder**: There is variability in the number of murders reported each year, with peaks observed in the mid-1990s and varying trends thereafter. The numbers have generally fluctuated without a clear upward or downward trend over the entire period.

- **Rape**: Shows an increase in reported incidents from the mid-1970s to the mid-1990s, followed by fluctuations and varying trends. The numbers have shown some decline in recent years compared to peaks observed in earlier decades.

- **Robbery**: Generally, there is a decrease in reported robberies from the peak in the late 1970s to the present day, with fluctuations along the way. This could reflect changes in law enforcement strategies and societal factors impacting crime.

- **Motor Vehicle Theft (M/V Theft)**: Shows a general decline from higher numbers in the late 1970s and 1980s to lower numbers in recent years. This decline may be influenced by improved vehicle security measures.

- **Other Property Crimes**: Overall, property crimes such as Larceny Theft and Burglary (B & E) have remained prevalent but with fluctuations. Larceny Theft, in particular, shows consistent high numbers throughout.

### Trends and Observations:
- **Violent Crimes vs. Property Crimes**: While violent crimes like Murder and Aggravated Assault show fluctuations, property crimes like Larceny Theft and Burglary (B & E) have remained consistently reported at high levels.
  
- **Recent Trends**: In recent years (2010-2020), there appears to be a decrease in some crime types like Robbery and Motor Vehicle Theft, possibly reflecting improved law enforcement tactics and societal changes.

### Conclusion:
The data suggests that while some types of crimes in Maryland have shown fluctuations, others like Larceny Theft and Burglary (B & E) have consistently remained prevalent over the years. Understanding these trends can help policymakers and law enforcement agencies focus on effective strategies to address crime prevention and public safety concerns in the state.

# 3. Geographical Analysis: Which jurisdictions have the highest and lowest crime rates? How do these rates compare to the state average?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Group data by jurisdiction and calculate average overall crime rate
grouped_data = crime_data.groupby('JURISDICTION').mean()

# Calculate state average crime rate
state_average_crime_rate = grouped_data['OVERALL CRIME RATE PER 100,000 PEOPLE'].mean()

# Append state average to the grouped data for plotting
grouped_data.loc['State Average'] = state_average_crime_rate

# Sort the data for better visualization
grouped_data_sorted = grouped_data.sort_values(by='OVERALL CRIME RATE PER 100,000 PEOPLE')

# Plotting the data
plt.figure(figsize=(14, 8))
grouped_data_sorted['OVERALL CRIME RATE PER 100,000 PEOPLE'].plot(kind='bar', color='skyblue')
plt.axhline(y=state_average_crime_rate, color='red', linestyle='--', label=f'State Average: {state_average_crime_rate:.2f}')
plt.title('Overall Crime Rate Per 100,000 People by Jurisdiction')
plt.ylabel('Overall Crime Rate Per 100,000 People')
plt.xlabel('Jurisdiction')
plt.xticks(rotation=90)
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Group data by jurisdiction and calculate average overall crime rate
grouped_data = crime_data.groupby('JURISDICTION').mean()

# Calculate state average crime rate
state_average_crime_rate = grouped_data['OVERALL CRIME RATE PER 100,000 PEOPLE'].mean()

# Identify jurisdictions with highest and lowest crime rates
highest_crime_rate_jurisdiction = grouped_data['OVERALL CRIME RATE PER 100,000 PEOPLE'].idxmax()
lowest_crime_rate_jurisdiction = grouped_data['OVERALL CRIME RATE PER 100,000 PEOPLE'].idxmin()

highest_crime_rate = grouped_data.loc[highest_crime_rate_jurisdiction, 'OVERALL CRIME RATE PER 100,000 PEOPLE']
lowest_crime_rate = grouped_data.loc[lowest_crime_rate_jurisdiction, 'OVERALL CRIME RATE PER 100,000 PEOPLE']

# Print the results
print(f'The jurisdiction with the highest crime rate is: {highest_crime_rate_jurisdiction} with a rate of {highest_crime_rate}')
print(f'The jurisdiction with the lowest crime rate is: {lowest_crime_rate_jurisdiction} with a rate of {lowest_crime_rate}')
print(f'The state average crime rate is: {state_average_crime_rate}')

# Visualize the data
# Bar plot to compare crime rates
plt.figure(figsize=(12, 6))
plt.bar(['State Average', highest_crime_rate_jurisdiction, lowest_crime_rate_jurisdiction],
        [state_average_crime_rate, highest_crime_rate, lowest_crime_rate],
        color=['blue', 'red', 'green'])
plt.title('Comparison of Crime Rates')
plt.ylabel('Overall Crime Rate Per 100,000 People')
plt.xlabel('Jurisdiction')
plt.grid(True)
plt.show()
```

Comparison and Insights

Baltimore City: The highest crime rate jurisdiction, almost double the state average, indicating significant crime challenges.

Worcester and Prince George's Counties: Also show high crime rates, needing focused law enforcement and community safety measures.

Garrett and Carroll Counties: With the lowest crime rates, these areas are relatively safer compared to the state average.




# 4.Population Correlation: Is there a correlation between population size and crime rates in different jurisdictions? Does a larger population necessarily mean higher crime rates?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Group data by jurisdiction and calculate average values
grouped_data = crime_data.groupby('JURISDICTION').mean()

# Scatter plot to visualize the relationship between population size and overall crime rate
plt.figure(figsize=(12, 6))
sns.scatterplot(x=grouped_data['POPULATION'], y=grouped_data['OVERALL CRIME RATE PER 100,000 PEOPLE'])
plt.title('Population Size vs Overall Crime Rate Per 100,000 People')
plt.xlabel('Population Size')
plt.ylabel('Overall Crime Rate Per 100,000 People')
plt.grid(True)
plt.show()

# Calculate the correlation coefficient
correlation_coefficient = grouped_data['POPULATION'].corr(grouped_data['OVERALL CRIME RATE PER 100,000 PEOPLE'])
print(f'The correlation coefficient between population size and overall crime rate is: {correlation_coefficient}')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Ensure necessary columns are available and group data by jurisdiction
crime_data_by_jurisdiction = crime_data.groupby('JURISDICTION').sum()

# Calculate crime rates per 100,000 people for each jurisdiction
crime_data_by_jurisdiction['CRIME_RATE'] = crime_data_by_jurisdiction['GRAND TOTAL'] / crime_data_by_jurisdiction['POPULATION'] * 100000

# Select relevant columns for correlation analysis
relevant_columns = ['POPULATION', 'MURDER', 'RAPE', 'ROBBERY', 'AGG. ASSAULT', 'B & E', 'LARCENY THEFT', 'M/V THEFT', 'CRIME_RATE']

# Calculate the correlation matrix
correlation_matrix = crime_data_by_jurisdiction[relevant_columns].corr()

# Print the correlation matrix
# print(correlation_matrix)

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix of Crime Data')
plt.show()
```

A correlation coefficient of 0.55 suggests a moderate positive correlation between population size and crime rates. This means that, generally, as the population size increases, the crime rates tend to increase as well, but the relationship is not perfectly linear.

# 5. Crime Rate Changes: Analyze the rate percent change for different crime types. Which crimes have seen the most significant changes (increases or decreases) over the years?

```python
import pandas as pd
import matplotlib.pyplot as plt



# List of columns for rate percent change per 100,000 people
rate_percent_change_columns = [
    'MURDER  RATE PERCENT CHANGE PER 100,000 PEOPLE',
    'RAPE RATE PERCENT CHANGE PER 100,000 PEOPLE',
    'ROBBERY RATE PERCENT CHANGE PER 100,000 PEOPLE',
    'AGG. ASSAULT  RATE PERCENT CHANGE PER 100,000 PEOPLE',
    'B & E RATE PERCENT CHANGE PER 100,000 PEOPLE',
    'LARCENY THEFT  RATE PERCENT CHANGE PER 100,000 PEOPLE',
    'M/V THEFT  RATE PERCENT CHANGE PER 100,000 PEOPLE'
]

# Calculate the average rate percent change for each crime type
average_rate_percent_changes = crime_data[rate_percent_change_columns].mean()

# Plot the average rate percent changes
plt.figure(figsize=(12, 6))
average_rate_percent_changes.plot(kind='bar')
plt.title('Average Rate Percent Change Per 100,000 People (1975-2020)')
plt.ylabel('Average Rate Percent Change (%)')
plt.xlabel('Crime Type')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()

# Plot the trends over the years
plt.figure(figsize=(12, 8))
for column in rate_percent_change_columns:
    plt.plot(crime_data.index, crime_data[column], label=column)

plt.title('Rate Percent Change Per 100,000 People Over the Years')
plt.ylabel('Rate Percent Change (%)')
plt.xlabel('Year')
plt.legend(loc='upper right')
plt.grid(True)
plt.show()

# Highlight the crimes with the most significant changes
significant_increases = average_rate_percent_changes.idxmax()
significant_decreases = average_rate_percent_changes.idxmin()

print(f'The crime with the most significant increase is: {significant_increases}')
print(f'The crime with the most significant decrease is: {significant_decreases}')
```

# 6. Crime Hotspots: Identify and visualize crime hotspots. Which areas should be prioritized for increased policing and resource allocation?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Filter data for specific years (2020, 2019, 2018, 2017)
years_to_plot = [2020, 2019, 2018, 2017]

for year in years_to_plot:
    # Filter data for the current year
    filtered_data = crime_data[crime_data['YEAR'] == year]
    
    # Aggregate crime rates by jurisdiction for the current year
    jurisdiction_crime = filtered_data.groupby('JURISDICTION')[['MURDER', 'RAPE', 'ROBBERY', 'AGG. ASSAULT', 'B & E', 'LARCENY THEFT', 'M/V THEFT']].mean()

    # Calculate total crime rate (sum of all crime types)
    jurisdiction_crime['TOTAL_CRIME_RATE'] = jurisdiction_crime.sum(axis=1)

    # Sort jurisdictions by total crime rate (descending)
    jurisdiction_crime_sorted = jurisdiction_crime.sort_values(by='TOTAL_CRIME_RATE', ascending=False)

    # Reset index to ensure proper x-axis ordering
    jurisdiction_crime_sorted = jurisdiction_crime_sorted.reset_index()

    # Define colors using a colormap (adjust cmap and vmin/vmax as needed)
    cmap = plt.cm.get_cmap('Oranges')  # Choose a colormap
    normalize = plt.Normalize(vmin=jurisdiction_crime_sorted['TOTAL_CRIME_RATE'].min(), vmax=jurisdiction_crime_sorted['TOTAL_CRIME_RATE'].max())
    colors = [cmap(normalize(value)) for value in jurisdiction_crime_sorted['TOTAL_CRIME_RATE']]

    # Plotting the total crime rate for visualization
    plt.figure(figsize=(12, 8))
    bars = plt.bar(jurisdiction_crime_sorted['JURISDICTION'], jurisdiction_crime_sorted['TOTAL_CRIME_RATE'], color=colors)

    # Add color bar indicating crime rate intensity
    sm = plt.cm.ScalarMappable(cmap=cmap, norm=normalize)
    sm.set_array([])
    cbar = plt.colorbar(sm)
    cbar.set_label('Total Crime Rate Intensity')

    plt.title(f'Total Crime Rate by Jurisdiction (Year: {year})')
    plt.xlabel('Jurisdiction')
    plt.ylabel('Total Crime Rate')
    plt.xticks(rotation=90)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

 Baltimore City, Prince George's County, Baltimore County, Montgomery County should be prioritized for increased policing and resource allocation
