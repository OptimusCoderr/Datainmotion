# Easy
Write a function that takes a string as input and returns the string reversed.

```python
input_string = "hello"

def backward(a):
    my_string = a
    reversed_string = my_string[::-1]
    return print(reversed_string)

backward(input_string)
```

# Medium
Write a function that takes a sentence and returns the longest word in the sentence.

```python
import re

input_sentence = "Data analysis is the key to success"

def longest(string):
    word_pattern = "\w+"

    regex = re.compile(word_pattern)
    words_found = regex.findall(string)

    if words_found:
        longest_word = max(words_found, key=lambda word: len(word))
    return print(longest_word)

longest(input_sentence)
```

# Hard
Write a function that finds all anagrams in a list of words and groups them together.


```python
def group_anagrams(words):
    anagram_dict = {}

    for word in words:
        sorted_word = ''.join(sorted(word))
        if sorted_word not in anagram_dict:
            anagram_dict[sorted_word] = [word]
        else:
            anagram_dict[sorted_word].append(word)

    # Extract the grouped anagrams from the dictionary
    grouped_anagrams = list(anagram_dict.values())
    
    return grouped_anagrams

words = ["listen", "silent", "enlist", "hello", "ohell"]
grouped_anagrams = group_anagrams(words)

for group in grouped_anagrams:
    print(group)
```
