# EASY

```python
def celsius_to_fahrenheit(ctof):
    temp = []
    for celsius in ctof:
        f_degrees = (celsius * 9/5) + 32
        temp.append(f_degrees)
    return temp

celsius_temperatures = [0, 20, 37, 100]
fahrenheit_temperatures = celsius_to_fahrenheit(celsius_temperatures)
print(fahrenheit_temperatures)

```

# MEDIUM

```python
def intersect(list1, list2):
    
    #The set function automatically removes duplicates and allows intersection while converting the list to a set
    set1 = set(list1)
    set2 = set(list2)
    
    set3 = set1.intersection(set2)
    
    return list(set3)

list1 = [1, 2, 3, 4, 5, 5, 6]
list2 = [4, 5, 6, 7, 8, 4]

result = intersect(list1,list2)
print(result)
```

# HARD

```python
def length_longest_substring(a):
    char_index_map = {}
    longest_length = 0
    start = 0
    substrings = []

    for end, char in enumerate(a):
        if char in char_index_map and char_index_map[char] >= start:
            start = char_index_map[char] + 1
        char_index_map[char] = end
        current_length = end - start + 1
        
        if current_length > longest_length:
            longest_length = current_length
            substrings = [a[start:end + 1]]
        elif current_length == longest_length:
            substrings.append(a[start:end + 1])

    # Print the substrings
    for substring in substrings:
        print(f"Substring: {substring}")

    return longest_length

a = "abrkaabcdefghijjxxx"
result = length_longest_substring(a)
print(result)
        
```

# BONUS OOP

```python
class Library:
    def __init__(self):
        self.books = set()

    def add_book(self, title):
        self.books.add(title)

    def remove_book(self, title):
        if title in self.books:
            self.books.remove(title)
        else:
            print(f"Book '{title}' not found in the library.")

    def search_book(self, title):
        return title in self.books

# Usage
library = Library()
library.add_book("The Great Gatsby")
library.add_book("1984")
library.add_book("To Kill a Mockingbird")

library = Library()
library.add_book("The Great Gatsby")
library.add_book("1984")
library.add_book("To Kill a Mockingbird")

print(library.search_book("1984")) 
print(library.search_book("The Catcher in the Rye")) 

library.remove_book("1984")
print(library.search_book("1984"))
```
